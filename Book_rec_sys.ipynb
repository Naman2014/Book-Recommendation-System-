# Collaborative Filtering Based Recommender System (USER BASED )

```python
# extracting users which gave 200+ ratings and it means they are regular users
x = ratings_new.groupby('User-ID').count()['Book-Rating'] > 200
reg_user = x[x].index
```

```python
filtered_rating = ratings_new[ratings_new['User-ID'].isin(reg_user)]
```

```python
# extracting books that have received more than 50 ratings and it means those are famous ones
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
# creating a pivot table that will have Book-title as index and user-id as columns and book rating as values
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index == book_name)[0][0]
    
    # sorting out top 4 similarity scores
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        book_titles = ['{} by {}'.format(title, author) for title, author in zip(temp_df.drop_duplicates('Book-Title')['Book-Title'].values, temp_df.drop_duplicates('Book-Title')['Book-Author'].values)]
        item.extend(book_titles)
        
        data.append(item)
    
    # Printing the books similar to the specified book
    print(f"Books similar to {book_name} are:")
    for item in data:
        print(item)
```

```python
# recommending a book by taking input
bk= input("Enter book name\n")
```

```python
recommend(bk)
```
